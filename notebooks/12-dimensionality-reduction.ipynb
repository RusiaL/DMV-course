```python
import pandas as pd

iris = pd.read_csv("data/12-iris.csv")

X = iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = iris.Species
```

## PCA realization
https://www.kaggle.com/code/vipulgandhi/pca-beginner-s-guide-to-dimensionality-reduction/notebook

```python
# Standartization
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
```

```python
# Computing Eigenvectors and Eigenvalues
import numpy as np

X_mean = np.mean(X, axis=0)
# cov_mat = np.cov(X)
cov_mat = (X - X_mean).T.dot((X - X_mean)) / (X.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)

eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('\nEigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
# Selecting Principal Components

# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort(key = lambda x: x[0], reverse=True)

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
    
print('\nExplained Variance')
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
print(cum_var_exp)
```

```python
# Projection matrix
matrix_w = np.hstack((eig_pairs[0][1].reshape(4,1),
                      eig_pairs[1][1].reshape(4,1)))

print('Matrix W:\n', matrix_w)
```

```python
# Projection Onto the New Feature Space
Y = X.dot(matrix_w)
```

```python
# Vizualization
import matplotlib.pyplot as plt

with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    for lab, col in zip(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'), 
                        ('blue', 'red', 'green')):
        plt.scatter(Y[y==lab, 0], Y[y==lab, 1], label=lab, c=col)
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(loc='lower center')
    plt.tight_layout()
    plt.show()
```

## PCA in sklearn library

```python
from sklearn.decomposition import PCA as sklearnPCA

sklearn_pca = sklearnPCA(n_components=2)
Y_sklearn = sklearn_pca.fit_transform(X)

sklearn_pca.explained_variance_ratio_
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    for lab, col in zip(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'),
                        ('blue', 'red', 'green')):
        plt.scatter(Y_sklearn[y==lab, 0], Y_sklearn[y==lab, 1],
                    label=lab, c=col)
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(loc='lower center')
    plt.tight_layout()
    plt.show()
```

## t-SNE (t-Distributed Stochastic Neighbour Embedding)

```python
from sklearn.manifold import TSNE

# perplexity is a hyperparameter
tsne_low_perplexity = TSNE(n_components=2, perplexity=5)
reduced_data_low_perplexity = tsne_low_perplexity.fit_transform(X)
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    for lab, col in zip(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'),
                        ('blue', 'red', 'green')):
        plt.scatter(reduced_data_low_perplexity[y==lab, 0], reduced_data_low_perplexity[y==lab, 1],
                    label=lab, c=col)
    plt.title('t-SNE with Low Perplexity')
    plt.legend(loc='lower center')
    plt.tight_layout()
    plt.show()
```

```python
tsne_high_perplexity = TSNE(n_components=2, perplexity=50)
reduced_data_high_perplexity = tsne_high_perplexity.fit_transform(X)
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))
    for lab, col in zip(('Iris-setosa', 'Iris-versicolor', 'Iris-virginica'),
                        ('blue', 'red', 'green')):
        plt.scatter(reduced_data_high_perplexity[y==lab, 0], reduced_data_high_perplexity[y==lab, 1],
                    label=lab, c=col)
    plt.title('t-SNE with Hight Perplexity')
    plt.legend(loc='lower center')
    plt.tight_layout()
    plt.show()
```

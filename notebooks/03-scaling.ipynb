```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
df = pd.DataFrame({
    'Age': [18, 22, 25, 30, 35, 60],
    'Income': [1000, 1500, 2000, 3000, 5000, 20000]  # є аутлайер
})

df
```

```python
scaler = StandardScaler()

df_scaled = df.copy()
df_scaled[['Age', 'Income']] = scaler.fit_transform(df[['Age', 'Income']])

df_scaled
```

```python
scaler = MinMaxScaler()

df_scaled = df.copy()
df_scaled[['Age', 'Income']] = scaler.fit_transform(df[['Age', 'Income']])

df_scaled
```

```python
scaler = RobustScaler()

df_scaled = df.copy()
df_scaled[['Age', 'Income']] = scaler.fit_transform(df[['Age', 'Income']])

df_scaled
```
